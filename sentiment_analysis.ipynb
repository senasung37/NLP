##뉴스기사 감정분석(Eng)

```python
#감정사전 라이브러리 설치
!pip install afinn
```

```python
#감정사전 라이브러리 불러오기
from afinn import Afinn
#데이터 라이브러리 불러오기
from sklearn.datasets import fetch_20newsgroups
```

```python
newsdata = fetch_20newsgroups(subset="train")
newsdata.data[4]
```

```python
#감정사전 구성, 데이터 감정가 계산
afinn = Afinn()
for i in range(10):
  print(afinn.score(newsdata.data[i]))
```

```python
#문장 만들기
sent1 = "It sucks! What the hack?"
sent2 = "It's wonderful. I like it very much. It is the best of the best!"
```

```python
#감정평가
print(afinn.score(sent1))
print(afinn.score(sent2))
```

```python
#긍정, 부정, 중립 감정 카운트

pos = 0
neut = 0
neg = 0

for i in newsdata.data:
  score = afinn.score(i)
  if score > 0:
    pos += 1
  elif score == 0:
    neut += 1
  else:
    neg += 1
```

```python
#시각화
import numpy as np
import matplotlib.pyplot as plt

plt.bar(np.arange(3), [pos, neut, neg])
plt.xticks(np.arange(3), ['pos', 'neut', 'neg'])
plt.show()
```
