```python
# 모듈 불러오기
from gtts import gTTS
from IPython.display import Audio
from IPython.display import display
```

```python
# 텍스트
mytext = '안녕하세요. 저는 성세은이라고 합니다.'
  
# 언어
language = 'ko'
  
# TTS 변환
my_sound = gTTS(text=mytext, lang=language, slow=False)
```

```python
# 오디오파일 저장
my_sound.save("welcome.mp3")
```

```python
# 오디오파일 플레이
wn = Audio("welcome.mp3", autoplay=True)
display(wn)
```
